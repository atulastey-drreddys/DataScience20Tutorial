```python
x = input("Enter your string 1")
y = input("Enter your string 2")
print(x)
print(y)
if len(x) == len(y):
    sort_x = sorted(x)
    print(sort_x)
    sort_y = sorted(y)
    print(sort_y)
    if sort_x == sort_y:
        print("Anagram")
    else:
        print("Not anagram")
        print("Invalid string")
else:
    print("Not anagram")
    print("Length Mismatch")
```
