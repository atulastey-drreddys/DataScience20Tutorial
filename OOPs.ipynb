```python
class Employee():
    def __init__(self,Id,Name,Designation):
        self.Id = Id
        self.Name = Name
        self.Designation = Designation
    def TCS(self):
        print(f"TCS Employee id {self.Id},Name {self.Name}, Designation {self.Designation}")
    def Infosys(self):
        print(f"Infosys Employee id {self.Id},Name {self.Name}, Designation {self.Designation}")    
tata = Employee(1,"Rohit","DataScientist")
tata.TCS()
Inf = Employee("INF2","Abhimanyu","Tester")
Inf.Infosys()
print(isinstance(tata,Employee))
print(isinstance(Inf,Employee))
#print(isinstance(Display,Employee))
```

```python
# Single level inheritance

class Parent:
    def Balram(self):
        print("He is the father of Adity")
class Child(Parent):
    def Aditya(self):
        print("Aditya is following the property of his father")

cc = Child()
cc.Aditya()
cc.Balram()

pp = Parent()
pp.Balram()
pp.Aditya()
```

```python
# Multilevel
class GrandParent():
    def GP(self):
        print("Grandparants class method")
class Parent(GrandParent):
    def Balram(self):
        print("He is the father of Adity")
class Child(Parent):
    def Aditya(self):
        print("Aditya is following the property of his father")

cc = Child()
cc.Aditya()
cc.Balram()
cc.GP()
```

```python
# Multiple inheritance
class GrandParent():
    def GP(self):
        print("Grandparants class method")
class Parent():
    def Balram(self):
        print("He is the father of Adity")
class Child(Parent,GrandParent):
    def Aditya(self):
        print("Aditya is following the property of his father")

cc = Child()
cc.GP()
cc.Aditya()
cc.Balram()
```

```python
# Hierarchical Inheritance
class Parent():
    def Balram(self):
        print("Grandparants class method")
class Son(Parent):
    def Aditya(self):
        print("Aditya is the son of Balram")
class Daughter(Parent):
    def Pooja(self):
        print("pooja is the daughter of Balram")

dd =Daughter()
dd.Balram()
dd.Pooja()
```

```python
# Hybrid inheritance
class GrandParent():
    def GP(self):
        print("Grandparants class method")
class Parent(GrandParent):
    def Balram(self):
        print("He is the father of Adity")
class Child(Parent):
    def Aditya(self):
        print("Aditya is following the property of his father")
class Parent():
    def Balram(self):
        print("He is the father of Adity")
class Child(Parent,GrandParent):
    def Aditya(self):
        print("Aditya is following the property of his father")
```

```python
# Encapsulation

class Employee:
    def Public(self):
        print("Public")
    def _Protected(self):
        print("This is protected")
    def __Private(self):
        print("THis is a private method")
        print("We can't access the data directly")  
emp = Employee()
emp.Public()
emp._Protected()
emp._Employee__Private()
```

```python
def Func(n):
    n = 0
    while n < 10:
        n += 1
        yield n
fun = Func(10)
for i in fun:
    print(i)
```

```python
def Outer(Func):
    def Inner():
        print("Before")
        Func()
        print("After")
    return Inner
@Outer 
def Display():
    print("This is Updated Output" )  
Display()
```

```python
def Upper_decorate(func):
    def wrapper():
        upper = func()
        return upper.upper()
    return wrapper

@Upper_decorate
def Greet():
    return "Hello python"
Greet()
```
