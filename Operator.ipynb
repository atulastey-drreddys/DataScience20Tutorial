# Comparison

```python
c = 689
v = 453
print(c>v)
print(c<v)
print(c>=v)
print(c<=v)
print(c==v)
print(c!=v)
```

Logical Operator
1) and
2) or
3) not

```python
x = int(input("Enter the value of x"))
y = int(input("Enter the value of y"))
z = int(input("Enter the value of z"))
print(x,y,z)
print(x > y and z > x)
print(z >= y and x == y)
print(y<z and z == x and x!=y)
```

```python
x = int(input("Enter the value of x"))
y = int(input("Enter the value of y"))
z = int(input("Enter the value of z"))
print(x,y,z)
print(x > y or z > x)
print(z >= y or x == y)
print(y<z or z == x or x!=y)
print(x == y and z != x and z < y)
```

```python
i = True
j = False
print(not i)
print(not j)
```

# Identity operator

```python
i = 68
j = 90
k = 90
print(id(i))
print(id(j))
print(id(k))
print(i is j)
print(j is k)
```
