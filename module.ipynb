```python
import datetime
```

```python
dt = datetime.datetime.today()
print(dt)
print(dt.year)
print(dt.month)
print(dt.date())
print(dt.hour)
print(dt.second)
print(dt.weekday())
```

```python
print(dt.strftime("%a"))
print(dt.strftime("%A"))
print(dt.strftime("%b"))
print(dt.strftime("%B"))
print(dt.strftime("%c"))
print(dt.strftime("%C"))
print(dt.strftime("%d"))
print(dt.strftime("%D"))
print(dt.strftime("%x"))
print(dt.strftime("%X"))
print(dt.strftime("%H"))
print(dt.strftime("%h"))
print(dt.strftime("%S"))
print(dt.strftime("%m"))
print(dt.strftime("%M"))
print(dt.strftime("%n"))
print(dt.strftime("%y"))
print(dt.strftime("%Y"))
print(dt.strftime("%Y-%M"))
print(dt.strftime("%Y-%m-%d"))
```
