```python
def Age_Cat(n):
    if n > 0 and n < 20:
        return 'Teen'
    elif n > 21 and n < 40:
        return "Adult"
    elif n > 41 and n < 75:
        return "Citizen"
    else:
        return "SR Citizen"
```

```python
cl_df['Age_Cat'] = cl_df['Age'].apply(Age_Cat)
cl_df
```
