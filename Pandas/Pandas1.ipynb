```python
from seaborn import load_dataset
iris = load_dataset('iris')
iris
```

```python
plt.figure(figsize=(4,4))
num_df = iris.select_dtypes(exclude='object')
sns.heatmap(num_df.corr(),
            annot=True,cbar=True,
            cmap='RdBu')
plt.xticks(rotation = 45,fontsize = 7)
plt.yticks(rotation = 45,fontsize = 7)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(iris['petal_width'])
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.boxenplot(iris['petal_width'])
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.violinplot(iris['petal_width'])
plt.show()
```
