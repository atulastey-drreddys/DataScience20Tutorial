```python
import re
```

```python
x = "Rohit Gulzar is a best captain of India."
pattern = 'Rohit Gulzar'
result = re.match(pattern,x)
print(result)
print(result.start())
print(result.end())
```

```python
x = "Rohit Gulzar is a best captain of India."
pattern = 'Gulzar'
result = re.match(pattern,x)
print(result)
print(result.start)
```

```python
x = "Rohit Gulzar is a best captain of India. Rohit is a very talented student"
pattern = 'Gulzar'
result = re.search(pattern,x)
print(result)
print(result.start())
print(result.end())
```

```python
x = "Rohit Gulzar is a best captain of India. Rohit is a very talented student"
pattern = 'Rohit'
result = re.findall(pattern,x)
print(result)
```

```python
x = "Rohit Gulzar is a best captain of India. Rohit is a very talented student. rohit is best player "
pattern = 'Rohit'
result = re.findall(pattern,x,re.IGNORECASE)
print(result)
```

```python
strr = "Hey my contact number is 492584238"
result = re.search('\d',strr)
print(result)
```

```python
strr = "Hey my contact number is 492584238"
result = re.search('\d+',strr)
print(result)
```

```python
strr = "Hey my contact number is 492584238"
result = re.findall('\d+',strr)
print(result)
```

```python
strr = "Hey my contact number is 492584238 and rohit contact 438597358"
result = re.findall('\d+',strr) # for matching digit
print(result)
```

```python
strr = "Hey my contact number is 492584238 and rohit contact 438597358"
result = re.findall('\D+',strr) # for matching not digit
print(result)
```

```python
strr = "Hey my contact number is 492584238 and rohit contact 438597358"
result = re.findall('\w+',strr) # wwordsr
print(result)
```

```python
strr = "Hey my contact number is 492584238 and rohit contact 438597358"
result = re.findall('\W+',strr) # whitespace
print(result)
```

```python
sent = "The cat sat on the mat"
result = re.findall(r'\bcat\b',sent) 
print(result)
```

```python
sent = "The cat sat on the mat catterpillercat"
result1 = re.findall(r'\Brcat\B',sent) 
result2 = re.findall(r"\bcat\b",sent)
print(result1)
print(result2)
```
