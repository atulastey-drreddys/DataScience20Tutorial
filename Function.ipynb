```python
# Positional Argument
def Addition(a,b,c):
    Add = a + b + c
    return Add
Addition(56,45,32)
```

```python
# Keyword Argument
def Addition(a,b,c):
    Add = a + b + c
    return Add
Addition(b = 56, a = 45,c =32)
```

```python
# Default Argument
def Addition(a,b,c = 40):
    Add = a + b + c
    return Add
print(Addition(56,45))
print(Addition(56,34,87))
```

```python
def Arb_Arg(*n):
    Sum = 0
    for i in n:
        Sum += i
    return Sum
Arb_Arg(435,5,345,75)
```

```python
def Even_Odd(n):
    if n % 2 == 0:
        print("Even")
    else:
        print("Odd")
Even_Odd(6)
Even_Odd(93)
Even_Odd(8674)
```

```python
def PrimeNumber(n):
    if n <= 1:
        return 1
    else:
        for i in range(2,n):
            if n % i == 0:
                print('Not prime number')
                break
        else:
            print("Prime Number")
PrimeNumber(9)
```

```python
def Fabonacci(n):
    a = 0
    b = 1
    for i in range(n):
        print(a)
        a,b = b, a + b
Fabonacci(10)
#Fabonacci(50)
```

```python
def Fact(n):
    if n <= 1:
        return 1
    else:
        return n * Fact(n-1)
Fact(5)
```

```python
def Counter(n):
    if n <= 1:
        return 1
    else:
        return n + Counter(n-1)
Counter(5)
```

# Scope

```python
def Func():
    x = 45 # Local variable
    print(x) # 

Func()
```

```python
x = 100 # Global Variabl
def Func():
    x = 45 # Local variable
    print(x) # 
Func()
print(x)
```

```python
x = 100 # Global Variabl
def Func():
    global x # update the global variable using global keyword
    x = 45 # Local variable
    print(x) # 
Func()
print(x)
```
