```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
x = np.arange(5,60,3)
print(x)
plt.plot(x)
plt.show()
```

```python
y = [5,2,6,6,5,8,9,7,3,4,5]
plt.figure(figsize=(5,2))
plt.plot(y)
plt.show()
```

```python
y = [5,2,6,6,5,8,9,7,3,4,5]
plt.figure(figsize=(5,2))
plt.plot(y,color = 'green',marker = "^",
         linestyle = 'dashdot',markersize = 10,
         markeredgecolor = 'black',linewidth = 2)
plt.show()
```

```python
y = [5,2,6,6,5,8,9,7,3,4,5]
z = [4,5,6,8,6,3,2,4,5,8,10]
plt.figure(figsize=(5,2))
plt.plot(y,color = 'green',marker = "^",
         linestyle = 'dashdot',markersize = 10,
         markeredgecolor = 'black',linewidth = 2,
         label = 'Y')
plt.plot(z,color = 'orange',marker = "*",
         linestyle = 'dashdot',markersize = 10,
         markeredgecolor = 'red',linewidth = 2,
         label = 'Z')
plt.legend()
plt.show()
```

```python
import seaborn as sns
y = [5,2,6,6,5,8,9,7,3,4,5]
z = [4,5,6,8,6,3,2,4,5,8,10]
plt.figure(figsize=(5,2))
sns.lineplot(y,label = "Y")
sns.lineplot(z,label = 'Z')
plt.title("Line Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid()
plt.show()
```

```python
y = [5,2,6,6,5,8,9,7,3,4,5]
plt.figure(figsize=(5,2))
plt.hist(y)
plt.show()
```

```python
y = [5,2,6,6,5,8,9,7,3,4,5]
plt.figure(figsize=(5,2))
sns.histplot(y)
plt.title("Hist Chart")
plt.xlabel("Groups")
plt.ylabel("Frequency")
plt.show()
```

```python
cat = ['a','c','b','d']
sales = [569,456,753,234]
plt.figure(figsize=(5,2))
plt.bar(x = cat,height=sales)
plt.show()
```

```python
cat = ['a','c','b','d']
sales = [569,456,753,234]
plt.figure(figsize=(5,2))
sns.barplot(x = cat,y=sales,hue=cat)
plt.show()
```

```python
cat = ['a','c','b','d']
sales = [569,456,753,234]
plt.figure(figsize=(5,3))
plt.pie(sales,labels=cat,
        startangle = 90,
        autopct="%.2f%%",
        explode = [0.1,0.1,0.1,0.1])
#plt.legend()
plt.show()
```

```python
x = np.linspace(10,20,40)
y = np.linspace(5,10,40)
plt.scatter(x = x, y = y)
plt.show()
```

```python
x = [2,3,5,6,3,3,6,7,6,4,4]
y = [7,3,5,7,4,8,4,6,8,7,2]
z = [9,4,6,7,12,3,12,5,3,6,5]
plt.figure(figsize=(5,3))
plt.scatter(x,y,color = '#c4c412')
plt.scatter(x,z,color = '#86dee3')
plt.xlabel("X Axis")
plt.ylabel('Y and Z')
plt.show()
```

```python
from seaborn import load_dataset
iris = load_dataset('iris')
```

```python
iris
```

```python
# Line Chart
plt.figure(figsize=(5,2))
plt.plot(iris['petal_length'])
plt.show()
```

```python
plt.figure(figsize=(7,2))
sns.lineplot(iris['petal_width'])
sns.lineplot(iris['petal_length'])
plt.show()
```

```python
plt.figure(figsize=(6,2))
sns.histplot(iris['petal_width'])
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.barplot(x = iris['species'],
            y = iris['sepal_length'])
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.barplot(x = iris['species'],
            y = iris['sepal_length'],
            hue = iris['species'])
plt.show()
```

```python
iris['species'].value_counts().index
```

```python
plt.figure(figsize=(4,3))
sns.countplot(iris['species'])
plt.show()
```

```python
plt.figure(figsize=(2,2))
plt.pie(iris['species'].value_counts().values,
        labels=iris['species'].value_counts().index,
        autopct="%0.2f%%")
plt.show()
```

```python
plt.figure(figsize = (5,3))
sns.scatterplot(x = 'sepal_length',
                y = 'petal_length',
                data = iris)
plt.show()
```

```python
plt.figure(figsize = (5,3))
sns.scatterplot(x = 'sepal_length',
                y = 'petal_length',
                hue = 'species',
                data = iris,)
plt.show()
```
