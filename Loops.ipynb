```python
x = ['bottle','mobile',2,4,6,4,7.8,45.64,"Mouse"]
flt  = []
intgr = []
Str = []
for i in x:
    if type(i) == str:
        Str.append(i)
    elif type(i) == int:
        intgr.append(i)
    else:
        flt.append(i)
print(flt)
print(intgr)
print(Str)
```

```python
num = int(input("Enter your number"))
print(num)
my_list = []
for i in range(1,num): # 1, 27
    if num % i== 0:
        my_list.append(i)
        Sum = 0
        for j in my_list:
            Sum += j
            if Sum == num:
                print("perfect number",Sum, num)
print("Factor list",my_list)
print(Sum)
```
